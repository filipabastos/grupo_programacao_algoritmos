```python
# Read the Players and Teams data into dataframes
f = open('Players.csv')
playersDF = pd.read_csv(f)
f = open('Teams.csv')
teamsDF = pd.read_csv(f)
```

```python
#Qual jogador (print do ultimo nome) de uma equipa com "ia" no nome, 
#que jogou menos de 200 minutos e fez mais de 100 passes?

jogadorPretendidoFatia = playersDF[playersDF.team.str.contains("ia") & (playersDF.minutes < 200) & (playersDF.passes > 100)]
jogador = jogadorPretendido.loc[431,"surname"] 
print("O nome do joagdor pretendido é " + jogador)
```

```python
# Qual é o número médio de passes feitos pelos avançados? Pelos meio-campistas?
# Verifique se a resposta especifica qual é qual e não inclua outras
# posições no seu resultado.

avancados = playersDF[playersDF.position == ("forward")]
meioCampistas = playersDF[playersDF.position == ("midfielder")]

numeroMedioAvancados = avancados.passes.mean()
numeroMedioMeioCampista = meioCampistas.passes.mean()

print("Número médio de passes feitos por avançados =", numeroMedioAvancados, "\nNúmero médio de passes feitos por meio-campistas =", numeroMedioMeioCampista)
```

```python
# Qual equipe tem a maior relação (razão, racio, etc) de goalsFor versus goalsAgainst?
# Imprima apenas o nome da equipa.
# Dica: adicione uma coluna "racio" ao dataframe das equipas e, depois, faça o seu "sort" e, 
# em seguida, use a head(1) ou tail(1) ...

teamsDF["Ratio"] = teamsDF["goalsFor"]/teamsDF["goalsAgainst"]
teamOrganizacao = teamsDF.sort_values(\
                        by = "Ratio",\
                        inplace = False,
                        ascending = False)

#teamOrganizacao.head() usado para consultar a equipa com maior relação
nomeDeEquipa = teamOrganizacao.loc[2,"team"]
print("A equipa que tem maior relação de goalsFor vs goalsAgainst é:",nomeDeEquipa)
```

```python
# Quantos jogadores que jogam numa equipa com ranking <10, 
# jogaram mais de 350 minutos?
# Lembrete: len() fornece o número de linhas que um dataframe contém

InfDez = teamsDF[teamsDF["ranking"] <10] 
equipasInfDez = InfDez.team
jogadoresDez = playersDF[playersDF.team.isin(equipasInfDez)]

jogadoresDez350min = jogadoresDez[(jogadoresDez.minutes > 350)] 
numeroJogadoresDez350min = len(jogadoresDez350min)
print("O número de jogadores que jogaram numa equipa com ranking <10, mais de 350 minutos foi:", numeroJogadoresDez350min, "jogadores")
```

```python
# Read the Titanic data into a dataframe
f = open('Titanic.csv')
titanicDF = pd.read_csv(f)
titanicDF
```

```python
# Quantas mulheres casadas embarcaram em Cherbourg?
# Nota: 'first' é uma função no Pandas, então 'titanic.first' irá gerar um erro;
# use 'titanic[' first ']

pessoasEmbarcamChergourg = titanicDF[titanicDF.embarked == ("Cherbourg")]
mulheresCasadas = pessoasEmbarcamChergourg[pessoasEmbarcamChergourg["first"].str.contains("Mrs.")]
numeroDeMulheresCasadas = len(mulheresCasadas)

print("O número de mulheres casadas que embarcou em Cherbourg foi:", numeroDeMulheresCasadas)
```

```python
# Liste a tarifa média paga pelos passageiros em cada uma das cidades de embarque.

embarquesEmSouthampton = titanicDF[titanicDF.embarked == ("Southampton")]
embarquesEmCherbourg = titanicDF[titanicDF.embarked == ("Cherbourg")]
embarquesEmQueenstown = titanicDF[titanicDF.embarked == ("Queenstown")]

south =  embarquesEmSouthampton.fare.mean()
cher = embarquesEmCherbourg.fare.mean()
queen = embarquesEmQueenstown.fare.mean()


print("Southampton = ", south , "\nCherbourg =", cher , "\nQueenstown =",  queen)
```

```python
# Os valores inexistentes (missing)  no Pandas são nulos (null), impressos como 'NaN' e
# a function pd.isnull(v) verifica se um valor é nulo.
# Que fração dos passageiros não tem valor para a idade?

valoresInexistentes = titanicDF.isnull().sum(axis = 0)
numeroDePassageiros = len(titanicDF)
numeroDePassageirosSemIdade = valoresInexistentes.age


percentagem = int(numeroDePassageirosSemIdade/numeroDePassageiros*100)

print(percentagem,"% dos passageiros não têm valor para a idade")
```
